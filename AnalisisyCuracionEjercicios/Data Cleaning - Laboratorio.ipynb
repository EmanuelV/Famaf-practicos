## 2.1. Etiquetas de variables/columnas: no usar caracteres especiales

```python
# helpful character encoding module
import chardet

import string

building_permits = pd.read_csv("New_Building_Permits.csv")
```

Analizamos el encoding que tiene el Dataset

```python
# look at the first ten thousand bytes to guess the character encoding
with open("New_Building_Permits.csv", 'rb') as rawdata:
    result = chardet.detect(rawdata.read(10000))

# check what the character encoding might be
print(result)
```

Leemos el dataset con el enconding correcto

```python
building_permits = pd.read_csv("New_Building_Permits.csv", encoding='ascii')
```

Vemos que aunque el método anterior informó que el dataset estaba en "ascii", lo mismo hay caracteres especiales, que no respetan la consigna

```python
building_permits.columns[building_permits.columns != building_permits.columns.str.extract(r'^(\w+)$')] 
```

```python
building_permits.columns = building_permits.columns.str.replace(' ', '_')
building_permits.head()
```

## 2.2. Tratar valores faltantes

Veamos cuantos valores nulos tenemos

```python
# get the number of missing data points per column
missing_values_count = building_permits.isnull().sum()

# look at the # of missing points in the first ten columns
# Filtramos solo aquellos que tienen más de 50000 registros nulos
missing_values_count[missing_values_count > 50000]
```

Dado que no podemos dropear tantos registros, escogemos reemplazarlos por 0

```python
building_permits = building_permits.fillna(0)
```

Y controlamos de nuevo que ya no quedan valores nulos

```python
missing_values_count = building_permits.isnull().sum()
missing_values_count[missing_values_count > 0]
```

## 2.3. Codificar variables

Las variables categóricas deben ser etiquetadas como variables numéricas, no como cadenas

Vemos que algunas de las variables categóricas ya tienen su variable numérica asociada, Ej: "Permit_Type_Definition" y "Permit_Type". 

No obstante ponemos en práctica los métodos explicados en las Notebooks de ejemplo, con la variable categórica "Current_Status":

```python
from sklearn import preprocessing
```

```python
#building_permits.describe(include='all')
building_permits.describe(include=['O'])
```

```python
#copiamos el dataset para aplicar los métodos
bp = building_permits
```

```python
column='Current_Status'
# Create a label (category) encoder object
le = preprocessing.LabelEncoder()
# Fit the encoder to the pandas column
le.fit(bp[column])
```

```python
# View encoder mapping
dict(zip(le.classes_,le.transform(le.classes_)))
```

```python
bp['Current_Status']
```

```python
# Apply the fitted encoder to the pandas column
bp[column] = le.transform(bp[column]) 
bp.head()
```

```python
bp['Current_Status']
```

```python
from sklearn.preprocessing import LabelBinarizer
other_column = 'Permit_Type_Definition'

lb = LabelBinarizer()

lb_results = lb.fit_transform(bp[other_column])
pd.DataFrame(lb_results, columns=((other_column + '_') + pd.Series(lb.classes_))).head(15)
```
