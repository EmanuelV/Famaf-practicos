```python
import folium
import pandas as pd
from folium.plugins import MarkerCluster
from folium.plugins import FastMarkerCluster
pasos = pd.read_csv('puestos_fronterizos.csv')
SF_COORDINATES = (-40.05 , -65.35 )

map = folium.Map(location=SF_COORDINATES, zoom_start=4)
marker_cluster = MarkerCluster().add_to(map)
FastMarkerCluster(data=list(zip(pasos.Latitud, pasos.Longitud)) ).add_to(map)


display(map)
```
